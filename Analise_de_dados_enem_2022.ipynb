# Analise dos dados socio economicos por notas

```python
#novo df
socio_economico = ['NU_INSCRICAO','TP_SEXO','NU_NOTA_CH', 'NU_NOTA_LC', 'NU_NOTA_REDACAO', 'NU_NOTA_CN', 'NU_NOTA_MT', 'Q001', 'Q002']
df_socio_economico = df.filter(items = socio_economico).count()
df_socio_economico
```

* Vendo a contagem de inscritos (250890) com as notas, podemos ver que que tivemos 71.158 inscritos que não comparecerão para realizar as provas no 1º dia e no 2º dia tivemos mais 10.208, totalizando 81.366 faltantes. 

```python
#retirada das linhas vazias do df
df_socio_economico = df.filter(items = socio_economico).dropna()
df_socio_economico
```

* Foram retiradas 82.036 linhas com valores nulos do df



```python
#criação de dicionario
dic_q1q2 = {'A':'A - Nunca estudou.', 
            'B':'B - Não completou a 4ª série/5º ano do Ensino Fundamental.', 
            'C':'C - Completou a 4ª série/5º ano, mas não completou a 8ª série/9º ano do Ensino undamental.', 
            'D':'D - Completou a 8ª série/9º ano do Ensino Fundamental, mas não completou o Ensino Médio.', 
            'E':'E - Completou o Ensino Médio, mas não completou a Faculdade.', 
            'F':'F - Completou a Faculdade, mas não completou a Pós-graduação.', 
            'G':'G - Completou a Pós-graduação.', 
            'H':'H - Não sei.',}
```

```python
#add os valores do dicionario que foi criado no df_socio_economico
df_socio_economico['NO_Q001'] = [dic_q1q2[resp] for resp in df_socio_economico.Q001]#pai
df_socio_economico['NO_Q002'] = [dic_q1q2[resp] for resp in df_socio_economico.Q002]#mae
df_socio_economico
```

```python
#escolaridade da pai por inscrição do participante
rel_ins_pai =df_socio_economico.filter(items = ['NU_INSCRICAO', 'NO_Q001']).groupby('NO_Q001').count().sort_values(by = 'NU_INSCRICAO', ascending = False)
rel_ins_pai
```

* A tabela mostra que o maior numero de inscrições e de participantes cujos pais não terminaram a faculdade.
* O menor quantitativo e de participantes cujo pai nunca estudou.
Podemos ver essa relação no grafico a seguir 

```python
df_socio_economico.filter(items = ['Q001', 'NU_INSCRICAO']).groupby('Q001').count().plot()
plt.xlabel('Grau de educação do pai')
plt.ylabel('Participantes')
plt.grid(True)
plt.title('Numero de inscrição por grau de educação do pai')
```

```python
#escolaridade da mae por inscrição do participante
rel_ins_mae =df_socio_economico.filter(items = ['NU_INSCRICAO', 'NO_Q002']).groupby('NO_Q002').count().sort_values(by = 'NU_INSCRICAO', ascending = False)
rel_ins_mae
```

* O numero de inscritos cuja mãe não completou a faculdade e maior que o dos pais. 
* Ha um aumento dos inscritos cujas mães entraram e não terminaram a pós graduação e o numero de mães que concluiram a pós graduação tbm e maior. 
* O numero de inscritos que não sabe a escolaridade das mães tbm e mais baixo que o dos pais
* Os dados representados pelo "Não sei" podem ser por não conhecer o pai ou mãe. 






```python
df_socio_economico.filter(items = ['Q002', 'NU_INSCRICAO']).groupby('Q002').count().plot()
plt.xlabel('Grau de educação da mãe')
plt.ylabel('Participantes')
plt.grid(True)
plt.title('Numero de inscrição por grau de educação da mãe')
```

```python
#notas de matematica com a escolaridade do mae 
rel_mat_pai = df_socio_economico.filter(items = ['NU_NOTA_MT', 'NO_Q001']).groupby('NO_Q001').count().sort_values(by = 'NU_NOTA_MT', ascending = False)
rel_mat_pai
```
